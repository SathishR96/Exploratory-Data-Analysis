# Top 10 cities with high literacy

```python
y=df.groupby('state_name',as_index=False).sum()
y1=y.nlargest(10,['literates_total'])
sns.set(rc={'figure.figsize':(20,21)})
sns.set(font_scale=1.5)
```

```python
sns.barplot(x='state_name',y='literates_total',data=y1)
```

# Explore sex ratio of different states. 

```python
m=df.groupby('state_name',as_index=False).mean()
print('\n')
print('Top 5  with maximum sex ratio ')
m=m.nlargest(5,'sex_ratio')
pd.DataFrame(m)
```

```python
sns.barplot(x='state_name',y='sex_ratio',data=m)
```

```python
print('\n')
print('Bottom 5  States of sex ratio ')
n=df.groupby('state_name',as_index=False).mean()
n=n.nsmallest(5,'sex_ratio')
n
```

```python
sns.barplot(x='state_name',y='sex_ratio',data=n)
```

# Percentage of total Graduate against Total Population of States. 

```python
f=df.groupby('state_name',as_index=False).sum()
f['percentage']=(f['total_graduates']/f['population_total'])*100
f=f[['state_name','percentage']]
f
```

```python
sns.set(font_scale=1)
sns.barplot(x='state_name',y='percentage',data=f)
```
